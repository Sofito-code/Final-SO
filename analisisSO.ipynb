#Escalabilidad

```python
for name, current_df in dfs.items():
    print(f"#### {name} - Escalabilidad\n")
    plt.figure(figsize=(14, 6))

    # Gráfico de Speedup vs. numProcesses
    plt.subplot(1, 2, 1)
    # *** MODIFICACIÓN: Usar 'matrixDimensions' para hue ***
    sns.lineplot(data=current_df, x='numProcesses', y='speedup', hue='matrixDimensions', marker='o', ci='sd')
    plt.title(f'{name}: Speedup vs. Number of Processes')
    plt.xlabel('Number of Processes')
    plt.ylabel('Speedup')
    plt.xscale('log')
    plt.grid(True, which="both", ls="--", c="0.7")
    plt.legend(title='Matrix Size')

    # Gráfico de Eficiencia vs. numProcesses
    plt.subplot(1, 2, 2)
    # *** MODIFICACIÓN: Usar 'matrixDimensions' para hue ***
    sns.lineplot(data=current_df, x='numProcesses', y='efficiency', hue='matrixDimensions', marker='o', ci='sd')
    plt.title(f'{name}: Efficiency vs. Number of Processes')
    plt.xlabel('Number of Processes')
    plt.ylabel('Efficiency (Speedup / Processes)')
    plt.xscale('log')
    plt.grid(True, which="both", ls="--", c="0.7")
    plt.legend(title='Matrix Size')

    plt.tight_layout()
    plt.show()
    print("\n" + "="*50 + "\n")
```

#Tamaño de Problema

```python
for name, current_df in dfs.items():
    print(f"#### {name} - Tamaño de Problema\n")
    # Convertir tiempos de ms a segundos
    current_df['timeParSec'] = current_df['timeParMs'] / 1000
    current_df['timeSeqSec'] = current_df['timeSeqMs'] / 1000

    plt.figure(figsize=(14, 6))

    # Gráfico de barras agrupadas: Tiempo (Secuencial y Paralelo) según tamaño de matriz
    plt.subplot(1, 2, 1)
    df_melted = current_df.melt(id_vars=['matrixDimensions', 'numProcesses'], value_vars=['timeSeqSec', 'timeParSec'],
                                var_name='Time Type', value_name='Time (seconds)')
    # *** MODIFICACIÓN: Usar 'matrixDimensions' para x ***
    sns.barplot(data=df_melted, x='matrixDimensions', y='Time (seconds)', hue='Time Type', palette={'timeSeqSec': 'blue', 'timeParSec': 'green'})
    plt.title(f'{name}: Sequential vs. Parallel Time by Matrix Size and Number of Processes')
    plt.xlabel('Matrix Size (Rows x Cols)')
    plt.ylabel('Time (seconds)')
    plt.xticks(rotation=45)
    plt.legend(title='Time Type')
    plt.grid(axis='y', ls='--', c='0.7')

    # Análisis de pendiente en escala log-log para complejidad O(n^3)
    plt.subplot(1, 2, 2)
    current_df['N'] = current_df['matrixRows'] # Asumiendo matrices cuadradas, N = matrixRows
    df_avg_time = current_df.groupby('N')[['timeSeqSec', 'timeParSec']].mean().reset_index()

    sns.scatterplot(data=df_avg_time, x='N', y='timeSeqSec', label='Sequential Time', color='blue')
    sns.scatterplot(data=df_avg_time, x='N', y='timeParSec', label='Parallel Time', color='green')
    # Ajustar el coeficiente para visualizar O(N^3)
    # Se puede calcular un coeficiente más preciso si se ajusta una regresión lineal en la escala log-log
    # Mantenemos 'N' en el eje x para el análisis log-log de complejidad
    plt.plot(df_avg_time['N'], 0.00000000001 * df_avg_time['N']**3, color='red', linestyle='--', label='$O(N^3)$ Reference')

    plt.xscale('log')
    plt.yscale('log')
    plt.title(f'{name}: Log-Log Plot of Time vs. Matrix Size (N)')
    plt.xlabel('Matrix Size (N)')
    plt.ylabel('Time (seconds)')
    plt.grid(True, which="both", ls="--", c="0.7")
    plt.legend()

    plt.tight_layout()
    plt.show()
    print("\n" + "="*50 + "\n")
```

#Sobrecarga de IPC

```python
for name, current_df in dfs.items():
    print(f"#### {name} - Sobrecarga de IPC\n")
    current_df['cpuTotalMs'] = current_df['cpuSysMs'] + current_df['cpuUserMs']
    current_df['cpuSysRatio'] = current_df['cpuSysMs'] / current_df['cpuTotalMs'] * 100
    current_df['cpuUserRatio'] = current_df['cpuUserMs'] / current_df['cpuTotalMs'] * 100

    # Filtrar para evitar divisiones por cero si cpuTotalMs es 0
    df_filtered_cpu = current_df[current_df['cpuTotalMs'] > 0].copy()

    if not df_filtered_cpu.empty:
        plt.figure(figsize=(12, 7))
        # *** MODIFICACIÓN: Usar 'matrixDimensions' para hue ***
        sns.barplot(data=df_filtered_cpu, x='numProcesses', y='cpuSysRatio', hue='matrixDimensions', ci='sd', palette='viridis')
        plt.title(f'{name}: System CPU Usage Ratio by Number of Processes and Matrix Size')
        plt.xlabel('Number of Processes')
        plt.ylabel('System CPU Usage (%)')
        plt.grid(axis='y', ls='--', c='0.7')
        plt.legend(title='Matrix Size')
        plt.show()

        # Gráfico de barras apiladas de CPU System vs CPU User
        df_cpu_melted = df_filtered_cpu.melt(id_vars=['numProcesses', 'matrixDimensions', 'RB'], value_vars=['cpuSysMs', 'cpuUserMs'],
                                             var_name='CPU Type', value_name='Time (ms)')

        plt.figure(figsize=(14, 7))
        sns.barplot(data=df_cpu_melted, x='numProcesses', y='Time (ms)', hue='CPU Type', ci='sd', palette='magma')
        plt.title(f'{name}: CPU System vs CPU User Time by Number of Processes and Matrix Size')
        plt.xlabel('Number of Processes')
        plt.ylabel('CPU Time (ms)')
        plt.grid(axis='y', ls='--', c='0.7')
        plt.legend(title='CPU Type')
        plt.show()
    else:
        print(f"No hay datos de CPU válidos para {name} para este análisis.")
    print("\n" + "="*50 + "\n")
```

```python
print("### Comparación entre Ubuntu y RaspberryPiOS (Pruebas estadísticas)\n")
# Usamos df original porque la comparación es entre los dos RBs
matrix_dimensions_unique = df['matrixDimensions'].unique()

print("\nResultados de T-tests (Speedup entre Ubuntu y RaspberryPiOS por tamaño de matriz):")
for dim in matrix_dimensions_unique:
    ubuntu_speedup = df[(df['RB'] == 'Ubuntu') & (df['matrixDimensions'] == dim)]['speedup']
    rpi_speedup = df[(df['RB'] == 'RaspberryPiOS') & (df['matrixDimensions'] == dim)]['speedup']

    if len(ubuntu_speedup) > 1 and len(rpi_speedup) > 1:
        # Se usa Welch's t-test (equal_var=False) porque no asumimos varianzas iguales.
        t_stat, p_val = stats.ttest_ind(ubuntu_speedup, rpi_speedup, equal_var=False)
        print(f"Matrix Size {dim}: T-statistic = {t_stat:.3f}, P-value = {p_val:.3f}")
    else:
        print(f"Matrix Size {dim}: No hay suficientes datos para realizar una prueba t-test para ambos RBs.")
print("\n" + "="*50 + "\n")
```
